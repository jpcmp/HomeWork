# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
tot_sc = school_data_complete['school_name'].nunique()
tot_st = school_data_complete['student_name'].count()
tot_bt = school_data_complete.groupby('school_name')['budget'].mean().sum()
avg_mt = school_data_complete['math_score'].mean()
avg_re = school_data_complete['reading_score'].mean()
avg_pa = (avg_mt + avg_re)/2
pct_mt = (school_data_complete[school_data_complete['math_score'] >= 70].count()['math_score'])/tot_st
pct_re = (school_data_complete[school_data_complete['reading_score'] >= 70].count()['reading_score'])/tot_st
df = pd.DataFrame({'Total Schools':[tot_sc],
                   'Total Students':[tot_st],
                   'Total Budget':[tot_bt],
                   'Average Math Score':[avg_mt],
                   'Average Reading Score':[avg_re],
                   '% Passing Math':[pct_mt],
                   '% Passing Reading':[pct_re],
                   '% Overall Passing Rate':[avg_pa]})
df.style.format({'Total Students':'{:,.0f}',
                 'Total Budget':'${:,.2f}',
                   'Average Math Score':'{:,.2f}',
                   'Average Reading Score':'{:,.2f}',
                   '% Passing Math':'{:,.2%}',
                   '% Passing Reading':'{:,.2%}',
                   '% Overall Passing Rate':'{:,.2f}'})
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
top_schools = school_data_complete.groupby('school_name').agg({'type': 'first', 'student_name':'count','budget':'first','math_score':'mean','reading_score':'mean'})
top_schools = top_schools.rename(columns={'type':'School Type','student_name':'Total Students','budget':'Total School Budget','math_score':'Average Math Score','reading_score':'Average Reading Score'})
top_schools['Per Student Budget'] = top_schools['Total School Budget']/top_schools['Total Students']
mid = top_schools['Per Student Budget']
top_schools.drop(columns=['Per Student Budget'], inplace=True)
top_schools.insert(3, 'Per Student Budget', mid)
pct_sc_mt = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['student_name'].count()
top_schools = pd.merge(top_schools,pct_sc_mt.to_frame(), on='school_name')
top_schools['% Passing Math'] = top_schools['student_name']/top_schools['Total Students']
top_schools = top_schools.drop(columns=['student_name'])
pct_sc_re = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['student_name'].count()
top_schools = pd.merge(top_schools,pct_sc_re.to_frame(), on='school_name')
top_schools['% Passing Reading'] = top_schools['student_name']/top_schools['Total Students']
top_schools = top_schools.drop(columns=['student_name'])
top_schools['% Overall Passing Rate'] = (top_schools['% Passing Math']+top_schools['% Passing Reading'])/2
spender_schools = top_schools
top_schools = top_schools.sort_values(by=['% Overall Passing Rate'],ascending=False)
top_schools.style.format({'Total Students':'{:,.0f}',
                 'Total School Budget':'${:,.2f}',
                 'Per Student Budget':'${:,.2f}',
                   'Average Math Score':'{:,.2f}',
                   'Average Reading Score':'{:,.2f}',
                   '% Passing Math':'{:,.2%}',
                   '% Passing Reading':'{:,.2%}',
                   '% Overall Passing Rate':'{:,.2%}'})
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
bottom_schools = top_schools.sort_values(by=['% Overall Passing Rate'],ascending=True)
bottom_schools.style.format({'Total Students':'{:,.0f}',
                 'Total School Budget':'${:,.2f}',
                 'Per Student Budget':'${:,.2f}',
                   'Average Math Score':'{:,.2f}',
                   'Average Reading Score':'{:,.2f}',
                   '% Passing Math':'{:,.2%}',
                   '% Passing Reading':'{:,.2%}',
                   '% Overall Passing Rate':'{:,.2%}'})
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
mt_9 = school_data_complete[school_data_complete['grade'] == '9th'].groupby('school_name')['math_score'].mean()
mt_10= school_data_complete[school_data_complete['grade'] == '10th'].groupby('school_name')['math_score'].mean()
mt_11 = school_data_complete[school_data_complete['grade'] == '11th'].groupby('school_name')['math_score'].mean()
mt_12 = school_data_complete[school_data_complete['grade'] == '12th'].groupby('school_name')['math_score'].mean()
avg_mt_sc = pd.merge(mt_9.to_frame(),mt_10.to_frame(),on='school_name')
avg_mt_sc = avg_mt_sc.rename(columns={'math_score_x':'9th','math_score_y':'10th'})
avg_mt_sc = pd.merge(avg_mt_sc,mt_11.to_frame(),on='school_name')
avg_mt_sc = pd.merge(avg_mt_sc,mt_12.to_frame(),on='school_name')
avg_mt_sc = avg_mt_sc.rename(columns={'math_score_x':'11th','math_score_y':'12th'})
avg_mt_sc
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
re_9 = school_data_complete[school_data_complete['grade'] == '9th'].groupby('school_name')['reading_score'].mean()
re_10= school_data_complete[school_data_complete['grade'] == '10th'].groupby('school_name')['reading_score'].mean()
re_11 = school_data_complete[school_data_complete['grade'] == '11th'].groupby('school_name')['reading_score'].mean()
re_12 = school_data_complete[school_data_complete['grade'] == '12th'].groupby('school_name')['reading_score'].mean()
avg_re_sc = pd.merge(re_9.to_frame(),re_10.to_frame(),on='school_name')
avg_re_sc = avg_re_sc.rename(columns={'reading_score_x':'9th','reading_score_y':'10th'})
avg_re_sc = pd.merge(avg_re_sc,re_11.to_frame(),on='school_name')
avg_re_sc = pd.merge(avg_re_sc,re_12.to_frame(),on='school_name')
avg_re_sc = avg_re_sc.rename(columns={'reading_score_x':'11th','reading_score_y':'12th'})
avg_re_sc
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
size_schools = school_data_complete.groupby('school_name').agg({'size': 'first', 'student_name':'count','math_score':'mean','reading_score':'mean'})
size_schools = size_schools.rename(columns={'student_name':'Total Students','math_score':'Average Math Score','reading_score':'Average Reading Score'})
size_schools = pd.merge(size_schools,pct_sc_mt.to_frame(), on='school_name')
size_schools['% Passing Math'] = size_schools['student_name']/size_schools['Total Students']
size_schools = size_schools.drop(columns=['student_name'])
size_schools = pd.merge(size_schools,pct_sc_re.to_frame(), on='school_name')
size_schools['% Passing Reading'] = size_schools['student_name']/size_schools['Total Students']
size_schools = size_schools.drop(columns=['student_name'])
size_schools['% Overall Passing Rate'] = (size_schools['% Passing Math']+size_schools['% Passing Reading'])/2
size_schools['School Size'] = pd.cut(size_schools["size"], size_bins, labels=group_names)
size_schools = size_schools.drop(columns=['Total Students','size'])
size_schools = size_schools.groupby('School Size').mean()
size_schools.style.format({'Average Math Score':'{:,.2f}',
                   'Average Reading Score':'{:,.2f}',
                   '% Passing Math':'{:,.2%}',
                   '% Passing Reading':'{:,.2%}',
                   '% Overall Passing Rate':'{:,.2%}'})
```
