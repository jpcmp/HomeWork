### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
purchase_data["SN"].nunique()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
nui = purchase_data["Item ID"].nunique()
avgp = purchase_data["Price"].mean()
nop = purchase_data["Purchase ID"].count()
tr = purchase_data["Price"].sum()
summary = pd.DataFrame({'Number of Unique Items':[nui],
                        'Average Price':[avgp], 
                        'Number of Purchases':[nop], 
                        'Total Revenue':[tr]})
summary.style.format({'Number of Unique Items':'{:,.0f}',
                        'Average Price':'${:,.2f}', 
                        'Number of Purchases':'{:,.0f}', 
                        'Total Revenue':'${:,.2f}'})
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#gender_tot= purchase_data.groupby('Gender').unique(['SN']).count()
gender_tot = purchase_data.groupby('Gender')['SN'].nunique()
gender_pct = purchase_data.groupby('Gender')['SN'].nunique()/purchase_data.groupby('Gender')['SN'].nunique().sum()
gender_demo = pd.merge(gender_tot.to_frame(),gender_pct.to_frame(),on='Gender')
gender_demo = gender_demo.rename(columns={"SN_x": "Total Count","SN_y":"Percentage of Players"})
gender_demo.style.format({'Percentage of Players': '{:,.2%}'.format})
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
summary = purchase_data.groupby('Gender').agg({'Purchase ID': 'count', 'Price': ['mean','sum']})
summary.columns = summary.columns.droplevel(0)
avg1 = purchase_data.groupby('SN').agg({'Gender': 'first', 'Price':'sum'})
avg2 = avg1.groupby('Gender')['Price'].mean()
summary = pd.merge(summary,avg2.to_frame(),on='Gender')
summary = summary.rename(columns={"count": "Purchase Count","mean":"Average Purchase Price","sum":"Total Purchase Value","Price":"Avg Total Purchase per Person"})
summary = summary.style.format({'Average Purchase Price': '${:,.2f}','Total Purchase Value': '${:,.2f}','Avg Total Purchase per Person': '${:,.2f}'})
summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
unique_data = purchase_data.groupby('SN')['Age'].last()
unique_df = unique_data.to_frame()
bins = [0, 9, 14, 19, 24, 29, 34, 39, unique_df["Age"].max()]
group_labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
unique_df["Age Range"] = pd.cut(unique_df["Age"], bins, labels=group_labels)
summary_agerange = unique_df.groupby('Age Range')['Age'].count()
summary_age = summary_agerange.to_frame()
summary_age['Percentage of Players'] = summary_age['Age'] / summary_age['Age'].sum()
summary_age = summary_age.rename(columns={"Age": "Total Count"})
summary_age = summary_age.style.format({'Percentage of Players': '{:,.2%}'})
summary_age
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data_age = purchase_data
purchase_data_age["Age Range"] = pd.cut(purchase_data_age["Age"], bins, labels=group_labels)
purchase_data_age = purchase_data_age.groupby('Age Range').agg({'Price': ['count', 'mean','sum']})
purchase_data_age.columns = purchase_data_age.columns.droplevel(0)
avga = purchase_data.groupby('SN').agg({'Age': 'first', 'Price':'sum'})
avga["Age Range"] = pd.cut(avga["Age"], bins, labels=group_labels)
avga = avga.groupby('Age Range')['Price'].mean()
summary_pa = pd.merge(purchase_data_age,avga.to_frame(),on='Age Range')
summary_pa = summary_pa.rename(columns={"count": "Purchase Count", "mean": "Average Purchase Price", "sum": "Total Purchase Value", "Price": "Avg Total Purchase per Person"})
summary_pa = summary_pa.style.format({"Average Purchase Price": '${:,.2f}', "Total Purchase Value": '${:,.2f}', "Avg Total Purchase per Person": '${:,.2f}'})
summary_pa
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spenders_data = purchase_data.groupby('SN').agg({'Price': ['count', 'mean','sum']})
spenders_data.columns = spenders_data.columns.droplevel(0)
spenders_data = spenders_data.rename(columns={"count": "Purchase Count", "mean": "Average Purchase Price", "sum": "Total Purchase Value"})
spenders_data = spenders_data.sort_values(by='Total Purchase Value',ascending=False)
spenders_data = spenders_data.style.format({"Average Purchase Price": '${:,.2f}', "Total Purchase Value": '${:,.2f}'})
spenders_data
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popitems_data = purchase_data.groupby(['Item ID','Item Name']).agg({'Price': ['count', 'mean','sum']})
popitems_data.columns = popitems_data.columns.droplevel(0)
popitems_data = popitems_data.rename(columns={"count": "Purchase Count", "mean": "Item Price", "sum": "Total Purchase Value"})
popitems_data = popitems_data.sort_values(by=['Purchase Count','Total Purchase Value'],ascending=False)
popitems_data_s = popitems_data.style.format({"Item Price": '${:,.2f}', "Total Purchase Value": '${:,.2f}'})
popitems_data_s
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profititems_data = popitems_data.sort_values(by=['Total Purchase Value'],ascending=False)
profititems_data = profititems_data.style.format({"Item Price": '${:,.2f}', "Total Purchase Value": '${:,.2f}'})
profititems_data
```
